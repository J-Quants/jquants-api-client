# リフレッシュトークンの書き込み

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/J-Quants/jquants-api-client-python/blob/master/examples/20220825-000-write-refresh_token.ipynb)

J-Quants API のリフレッシュトークンを Google Drive の MyDrive/drive_ws/secret/jquantsapi-key.txt に書き込みます。

リフレッシュトークンは取得方法は以下のリンク先をご参照ください。

https://jpx.gitbook.io/j-quants-api/api-reference#idtkunno


----

**このノートブックはGoogle Driveを使用します。**

- Google Drive の以下のファイルにリフレッシュトークンを書き込みます。
    - `MyDrive/drive_ws/secret/jquantsapi-key.txt`


```python
# Google drive をマウントします
from google.colab import drive
drive.mount('/content/drive')

# 必要なモジュールを読み込みます
import os
from getpass import getpass


# refresh_token を入力します
refresh_token = getpass("Please input refresh token")

# refresh_token の書き込み先を指定します
OUTPUT_DIR = "/content/drive/MyDrive/drive_ws/secret"
OUTPUT_PATH = f"{OUTPUT_DIR}/jquantsapi-key.txt"

# refresh_token を書き込みます
os.makedirs(OUTPUT_DIR, exist_ok=True)
with open(OUTPUT_PATH, mode="w") as f:
  f.write(refresh_token)

# 完了文字列を表示します
print("Done")
```
